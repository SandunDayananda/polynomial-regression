```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate random data for the independent variable
np.random.seed(0)
X = np.random.rand(30) * 10

# Generate the dependent variable (target) using the equation
y = 2 + 3*X - 0.5*X**2 + np.random.randn(30) * 2

# Transform the independent variable using polynomial features
poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Generate the x values for plotting the curve
x_vals = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_vals_poly = poly.transform(x_vals)
y_pred = model.predict(X_vals_poly)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot the polynomial regression curve
plt.plot(x_vals, y_pred, color='red', label='Polynomial Regression Curve')

# Set the axis labels and title
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Polynomial Regression')

# Set the legend
plt.legend()

# Display the plot
plt.show()
```
