```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate random data for the independent variable
np.random.seed(0)
X = np.random.rand(30) * 10

# Generate the dependent variable (target) using a polynomial equation
y = 2 + 3*X - 0.5*X**2 + np.random.randn(30) * 2

# Transform the input variable using polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Generate the x values for plotting the curve
x_vals = np.linspace(X.min(), X.max(), 100)
X_mesh = poly.transform(x_vals.reshape(-1, 1))
y_pred = model.predict(X_mesh)

# Generate the x values for plotting the regression line
X_regression = poly.transform(x_vals.reshape(-1, 1))
y_regression = model.predict(X_regression)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(x_vals, y_regression, color='red', label='Regression Line')

# Set the axis labels and title
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Polynomial Regression')

# Set the legend
plt.legend()

# Display the plot
plt.show()
```
